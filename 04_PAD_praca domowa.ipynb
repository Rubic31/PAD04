```python
df['tshirt_category'] = df['tshirt_category'].apply(lambda x: x.replace('Wh',"White") if x.startswith("Wh ") else x.replace('Bl',"Black") if x.startswith("Bl ") else x)
```

```python
df
```

```python
df['tshirt_category'].unique()
```

```python
df['tshirt_category'].nunique()
```

```python
df['tshirt_category'] = df['tshirt_category'].apply(lambda x: x.replace('Tsh', "T-Sh") if 'Tshirt' in x else x)
```

```python
df['tshirt_category'].unique()
```

```python
df['tshirt_category'].nunique()
```

```python
#4 do rozdzielenia wystarczy split, nie potrzebny apply
df[['tshirt_colour', 'tshirt_type', 'tshirt_size']] = df.tshirt_category.str.split(expand=True)
```

```python
df
```

```python
df['tshirt_type'] = df['tshirt_colour'].apply(lambda x: x if (x.startswith("Hoo") or x.startswith("Tenni")) else 'T-Shirt')
```

```python
df['tshirt_colour'].loc[(df['tshirt_colour'] == 'Hoodie') | (df['tshirt_colour'] == 'Tennis')] = np.nan
```
